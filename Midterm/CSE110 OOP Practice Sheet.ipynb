```python
#Write your code here

# TODO
class Student:
    def __init__(self,n,i,e,g):
        self.name=n
        self.id=i
        self.enrolled=e
        self.gpa=g
    def calculate_CGPA(self):
        counter=0
        summ=0
        credit=3
        for i in range(len(self.gpa)):
            counter+=1
            summ+=self.gpa[i]*credit
            total_credit=credit*counter
        self.cgpa=summ/total_credit
    def print_details(self):
        print('---------','Department:',self.enrolled,'---------')
        print('Name:',self.name,' ','ID:',self.id)
        print('CGPA:',self.cgpa)
        if(3.80<=self.cgpa<=4.00):
            print("Your academic standing is 'Highest Distinction' ")
        if(3.65<=self.cgpa<=3.79):
            print("Your academic standing is 'High Distinction' ")
        if(3.50<=self.cgpa<=3.64):
            print("Your academic standing is 'Distinction' ")
        if(2<=self.cgpa<=3.49):
            print("Your academic standing is 'Satisfactory' ")
        if(0<=self.cgpa<1.99):
            print('You cannot graduate')
#Do not change the following code:
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
s1.print_details()

s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
s2.print_details()

s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
s3.print_details()
```
